```
import pandas as pd
import numpy as np
```

```
data = pd.read_excel("20250521165248.xlsx", engine="openpyxl")
```

```
data=data[data['執業狀態']=='執業']
data_本國=data[data['國籍']=='Taiwan, Republic of China台灣']
data_外籍=data[data['國籍']!='Taiwan, Republic of China台灣']
df_本國=pd.crosstab(data_本國['性別'],data_本國['縣市別']).T
df_本國=df_本國.reset_index()
df_本國=df_本國[['縣市別','男','女']]
data_外籍=pd.crosstab(data_外籍['性別'],data_外籍['縣市別']).T
data_外籍=data_外籍.reset_index()
data_外籍=data_外籍[['縣市別','男','女']]
df=df_本國.merge(data_外籍,on='縣市別',how='left')
df.columns=['縣市別', '本籍_男', '本籍_女', '外籍_男', '外籍_女']
df=df.fillna(0)
df['外籍_男']=df['外籍_男'].astype(int)
df['外籍_女']=df['外籍_女'].astype(int)
df['排序']=df['縣市別']
排序={'基隆市':1,'臺北市':2,'新北市':3,'桃園市':4,'新竹市':5,'新竹縣':6,'苗栗縣':7,'臺中市':8,'南投縣':9,'彰化縣':10,'雲林縣':11,'嘉義市':12,'嘉義縣':13,'臺南市':14,'高雄市':15,'屏東縣':16,'宜蘭縣':17,'花蓮縣':18,'臺東縣':19,'澎湖縣':20,'金門縣':21,'連江縣':22}
df['排序']=df['排序'].map(排序)
df=df.sort_values(by='排序')
df.index=range(len(df))
df=df.drop(columns='排序')
df.insert(3,'本籍合計',df['本籍_男']+df['本籍_女'])
df.insert(6,'外籍合計',df['外籍_男']+df['外籍_女'])
```
